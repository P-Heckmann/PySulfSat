```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import PySCSS as ss
import Thermobar as pt
pd.options.display.max_columns = None
```

```python
Kiseeva_in=pd.read_excel('Kiseeva_Kds.xlsx', sheet_name='PySCSS_testing')
```

```python
Calc_Kd=ss.calculate_Kiseeva_sulf_comp_kd(Ni_Sulf=Kiseeva_in['Ni_Sulf'],
                                    Cu_Sulf=Kiseeva_in['Cu_Sulf'], 
                                       FeOt_Liq=Kiseeva_in['FeOt_Liq'], 
                                       T_K=Kiseeva_in['T_K'],  
                                       Fe3Fet_Liq=0)
```

```python
Calc_Kd.head()
```

```python
fig, ((ax0, ax1), (ax2, ax3), (ax4, ax5)) = plt.subplots(3,2, figsize = (10,15)) # adjust dimensions of figure here
ax0.plot([np.min(Kiseeva_in['DNi']), np.max(Kiseeva_in['DNi'])],
        [np.min(Kiseeva_in['DNi']), np.max(Kiseeva_in['DNi'])],
         '-k')
ax0.plot(Kiseeva_in['DNi'], Calc_Kd['DNi'], 'ok', mfc='red')
ax0.set_xlabel('D_Ni Kiseeva Spreadsheet')
ax0.set_ylabel('D_Ni PySCSS')

ax1.plot([np.min(Kiseeva_in['DCu']), np.max(Kiseeva_in['DCu'])],
        [np.min(Kiseeva_in['DCu']), np.max(Kiseeva_in['DCu'])],
         '-k')
ax1.plot(Kiseeva_in['DCu'], Calc_Kd['DCu'], 'ok', mfc='red')
ax1.set_xlabel('D_Cu Kiseeva Spreadsheet')
ax1.set_ylabel('D_Cu PySCSS')

ax2.plot([np.min(Kiseeva_in['DAg']), np.max(Kiseeva_in['DAg'])],
        [np.min(Kiseeva_in['DAg']), np.max(Kiseeva_in['DAg'])],
         '-k')
ax2.plot(Kiseeva_in['DAg'], Calc_Kd['DAg'], 'ok', mfc='red')
ax2.set_xlabel('D_Ag Kiseeva Spreadsheet')
ax2.set_ylabel('D_Ag PySCSS')

ax3.plot([np.min(Kiseeva_in['DPb']), np.max(Kiseeva_in['DPb'])],
        [np.min(Kiseeva_in['DPb']), np.max(Kiseeva_in['DPb'])],
         '-k')
ax3.plot(Kiseeva_in['DPb'], Calc_Kd['DPb'], 'ok', mfc='red')
ax3.set_xlabel('D_Pb Kiseeva Spreadsheet')
ax3.set_ylabel('D_Pb PySCSS')


ax4.plot([np.min(Kiseeva_in['DZn']), np.max(Kiseeva_in['DZn'])],
        [np.min(Kiseeva_in['DZn']), np.max(Kiseeva_in['DZn'])],
         '-k')
ax4.plot(Kiseeva_in['DZn'], Calc_Kd['DZn'], 'ok', mfc='red')
ax4.set_xlabel('D_Zn Kiseeva Spreadsheet')
ax4.set_ylabel('D_Zn PySCSS')

ax5.plot([np.min(Kiseeva_in['DCd']), np.max(Kiseeva_in['DCd'])],
        [np.min(Kiseeva_in['DCd']), np.max(Kiseeva_in['DCd'])],
         '-k')
ax5.plot(Kiseeva_in['DCd'], Calc_Kd['DCd'], 'ok', mfc='red')
ax5.set_xlabel('D_Cd Kiseeva Spreadsheet')
ax5.set_ylabel('D_Cd PySCSS')

fig.tight_layout()
```
